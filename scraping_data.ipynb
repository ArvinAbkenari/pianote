```python
import re
def preprocess(text):
    text = re.sub(r'[^\w\s/]', '', str(text).lower())
    tokens = re.split(r'[\s/]+', text)
    return [token for token in tokens if token and token != 'or']

data['description_array'] = data['description'].apply(preprocess)
```

```python
data
```

```python
def extract_finish(tokens):
    return [t for t in tokens if t in {'satin', 'polish'}]

data['finish'] = data['description_array'].apply(extract_finish)
```

```python
data
```

```python
def remove_finish(tokens):
    return [t for t in tokens if t not in {'satin', 'polish'}]

data['description_array'] = data['description_array'].apply(remove_finish)
```

```python
data
```

```python
def remove_last_word(text):
    if pd.isna(text):
        return text
    return text.rsplit(' ', 1)[0] if ' ' in text else text

data['description_array_color'] = data['description'].apply(remove_last_word)
data
```

```python
def rePreprocess(text):
    if pd.isna(text):
        return text
    text = str(text).lower()
    text = re.sub(r'\sor\s', '/', text)     
    text = text.replace(',', '/')          
    text = re.sub(r'[^\w\s/]', '', text) 
    tokens = text.split('/')
    return [token.strip() for token in tokens if token.strip()]

data['description_array_color_x'] = data['description_array_color'].apply(rePreprocess)
data.head(20)
```

```python
main_colors = ['ebony' , 'mahogany' , 'white' , 'walnut' , 'cherry' , 'bubinga' , 'vavona' , 'yew' , 'rosewood' , 'chrome' , 'pommele' ,'black',
               'nickel' , 'oak' , 'makassar' , 'burl' , 'macassar' ,'extraordinary veneers' , 'special veneers' ,'ordinary veneers' ,'maple']
def change_color(text):
    if isinstance(text, float) and pd.isna(text):  
        return text
    
    if isinstance(text, list):
        updated_text = []
        for item in text:
            updated_item = item  
            for color in main_colors:
                if color in item:  
                    updated_item = color  
                    break  
            updated_text.append(updated_item)
        return updated_text
    return text
data['color_array_x'] = data['color_array'].apply(change_color)
data.head(10)
```
